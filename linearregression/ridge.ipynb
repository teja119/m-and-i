```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial import Voronoi , voronoi_plot_2d
import os
os.chdir("D:/m and i")
```

```python
data = pd.read_excel("linearregression/house_price.xlsx")
data.head()
```

```python
x = np.array(data[["X:House Size(100s ft2)"]])
y = np.array(data[["Y:Actual Selling Price"]])
```

```python

fig = voronoi_plot_2d(Voronoi(np.column_stack((x,y))))
plt.plot()
```

```python
m,n = x.shape
alpha = 0.0001
ridge_lambda = 0.01 
epoch = 1000
w = np.zeros((n,1))
bias = 0 
cost_history = []
```

```python
for i in range(epoch):
    y_pred = x.dot(w) + bias
    cost = y_pred - y 
    mean_error = (np.sum(cost**2))/2*m + ridge_lambda*w**2
    dw  = x.T.dot(cost) + ridge_lambda*2*w
    db = np.sum(cost)

    w-= alpha*dw
    bias-= alpha*db
    cost_history.append(mean_error)
```

```python
print(w,bias)
```

```python
y_line = x.dot(w)+bias
```
