```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
os.chdir("D:/m and i")
```

```python
data = pd.read_excel("linearregression/house_price.xlsx")
data.head()
```

```python
x= np.array(data[["X:House Size(100s ft2)"]])
y = np.array(data[["Y:Actual Selling Price"]])
```

```python
m , n = x.shape
w=np.zeros((n,1))
bias = 0 
lasso_lamda = 0.001 
alpha = 0.0001
tolerence = 1e-6
epochs = 1000
cost_history = []
```

```python
for i in range(epochs):
    y_pred = x.dot(w)+bias
    cost = y_pred - y 
    mean_error = (np.sum(cost**2))/(2*m) + lasso_lamda*np.sum(abs(w))
    dw  = x.T.dot(cost) + np.sign(w)*lasso_lamda
    db  = np.sum(cost) 
    cost_history.append(mean_error)
    w -= alpha*dw
    bias -= alpha*db

```

```python
y_line = x.dot(w)+bias
```

```python
plt.scatter(x , y , color = "red" , marker = "x")
plt.plot(x,y_line , color = "blue")
plt.xlabel("house size")
plt.ylabel("house price")
plt.legend()
plt.show()
```
